```python
# importinng dependencies
import pandas as pd
```

```python
# import Sample Table 1
sample_review_df = pd.read_csv("tables/sample_table_hotel_reviews.csv")
sample_review_df 
```

# Merging positive and negative review columns into 1 column

```python
# changing the No Negative to null
sample_review_df.loc[(sample_review_df["Negative_Review"] == "No Negative"), "Negative_Review"] = ''

# changing the No Positive to null
sample_review_df.loc[(sample_review_df["Positive_Review"] == "No Positive"), "Positive_Review"] = ''

sample_review_df
```

```python
# merge positve and negative review columns into a 'Combined review' column
sample_review_df["Combined_Review"] = sample_review_df["Negative_Review"] + sample_review_df["Positive_Review"]

sample_review_df
```

```python
# export as csv
sample_review_df.to_csv('sample_table_for_NLP.csv', index=False)
```
