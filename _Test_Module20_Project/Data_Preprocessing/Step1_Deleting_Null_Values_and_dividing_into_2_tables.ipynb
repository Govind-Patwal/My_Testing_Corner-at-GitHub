```python
# importing dependencies
import pandas as pd
import zipfile as zf
```

```python
# importing file and converting to a df
zip_file = zf.ZipFile('../DataFiles/RawData/archive.zip') 
hotel_df = pd.read_csv(zip_file.open('Hotel_Reviews.csv'))

# since there is no unique column in the table, we will use the index to creae a unique column for working with data in PostgreSQL
hotel_df.index.name='Review_Id'

# check the df
hotel_df 
```

```python
# check for null/na values
hotel_df.isnull().sum() 
```

```python
# drop rows that are na/null and reset the index from 0
hotel_df = hotel_df.dropna().reset_index(drop=True) 

# the operation above will delete the index name, so renaming it again
hotel_df.index.name='Review_Id'

hotel_df 
```

```python
# check for null again
hotel_df.isnull().sum() 
```

# Breaking the big table into 2 tables: 1st with hotel reviews, 2nd with hotel locations

```python
## Table 1: Hotel reviews only - no location data

# hotel reviws without lat and lng
hotel_reviews = hotel_df.drop(columns=['lat','lng'])
hotel_reviews
```

```python
## Table 2: DF with locations (no reviews)

# creating a subset of the dataframe with unique values on the column 'Hotel_Address' and then reset the index to 0
unique_address_df = hotel_df.drop_duplicates(subset=['Hotel_Address']).reset_index(drop=True) 
unique_address_df
```

```python
# selecting ONLY 4 columns
hotel_location = unique_address_df[["Hotel_Address", "Hotel_Name", "lat", "lng"]]
hotel_location
```

# Exporting as zip files

```python
# Exporting hotel_reviews as a Zip file
compression_opts = dict(method='zip', archive_name='hotel_review.csv')  
hotel_reviews.to_csv('../DataFiles/TransformedData/tables/hotel_reviews.zip', index=True, compression=compression_opts)  
```

```python
# Exporting hotel_location as a Zip file
compression_opts = dict(method='zip', archive_name='hotel_location.csv')  
hotel_location.to_csv('../DataFiles/TransformedData/tables/hotel_location.zip', index=False, compression=compression_opts)  
```
