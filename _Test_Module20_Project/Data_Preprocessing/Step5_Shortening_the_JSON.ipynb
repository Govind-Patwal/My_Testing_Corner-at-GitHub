```python
# importing dependencies
import pandas as pd
import zipfile as zf
```

# Reading the JSON as a DF

```python
# Reading the JSON to make sure it was exported correctly
zip_file = zf.ZipFile('../DataFiles/TransformedData/JSON/hotel_combined.zip') 
df = pd.read_json(zip_file.open('hotel_combined.json'))

# check the df
df
```

```python
# selecting 10,000 rows
thousand_row_JSON = df.loc[0:9999]
thousand_row_JSON
```

# Exporting JSON as a ZIP file

```python
# Exporting hotel_combined.json as a Zip file
compression_opts = dict(method='zip', archive_name='thousand_row_JSON.json')  
thousand_row_JSON.to_json('../DataFiles/TransformedData/JSON/thousand_row_JSON.zip', index=True, compression=compression_opts) 
```
