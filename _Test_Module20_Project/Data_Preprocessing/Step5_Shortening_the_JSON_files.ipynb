```python
# importing dependencies
import pandas as pd
import zipfile as zf
```

# Reading the JSON as a DF

```python
# Reading the JSON Zip file that has over 500,000 rows of combined data
zip_file = zf.ZipFile('../DataFiles/TransformedData/JSON/all_rows_JSON.zip') 
df = pd.read_json(zip_file.open('all_rows_JSON.json'))

# check the df
df
```

# Selecting 10,000 rows and exporting as a JSON

```python
# selecting 10,000 rows
ten_thousand_row_JSON = df[0:10000]
ten_thousand_row_JSON
```

```python
# Exporting 10,000 hotel_combined.json as a Zip file
compression_opts = dict(method='zip', archive_name='ten_thousand_row_JSON.json')  
ten_thousand_row_JSON.to_json('../DataFiles/TransformedData/JSON/ten_thousand_row_JSON.zip', index=True, compression=compression_opts) 
```

```python
# Reading the JSON to make sure it was exported correctly
zip_file = zf.ZipFile('../DataFiles/TransformedData/JSON/ten_thousand_row_JSON.zip') 
df_10k = pd.read_json(zip_file.open('ten_thousand_row_JSON.json'))

# check the df
df_10k
```
