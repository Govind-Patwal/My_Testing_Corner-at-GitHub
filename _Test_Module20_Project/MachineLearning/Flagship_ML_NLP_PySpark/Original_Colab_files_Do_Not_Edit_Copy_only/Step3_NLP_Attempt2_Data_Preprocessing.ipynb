```python
# importing the table created in the last NLP Attempt (Attempt 1)
hotel_reviews = pd.read_csv('../../DataFiles/TransformedData/tables/hotel_reviews_for_NLP1_float_review_score.csv')
hotel_reviews
```

# Challenge 
* Choosing the format of Reviewer_Score as float resulted in too big an output range for the PySpark Algorithm, which resulted in very low accuracy of the model: 0.072433

# Solution: 
* Change the format of Reviewer_Score to int

```python
hotel_reviews['Reviewer_Score'] = hotel_reviews['Reviewer_Score'].astype(np.int64)
hotel_reviews
```

```python
# Exporting as a CSV file
hotel_reviews.to_csv('../../DataFiles/TransformedData/tables/hotel_reviews_for_NLP2_int_review_score.csv', index=False)
```
