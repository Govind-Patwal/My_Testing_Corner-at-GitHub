```python
# importing pandas as pd
import pandas as pd

# importing matplotlib as plt
import matplotlib.pyplot as plt
%matplotlib inline # magic command
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot
plt.plot(x_axis, y_axis)

#show the plot
plt.show()
```

```python
x_axis = [12,15,17,21,45]
y_axis = [34,52,34,45,34]
```

```python
# size of the plot, shiould be the first line
plt.figure(figsize = (30, 30))

# plot a line chart, and legend label as GSP
plt.plot(x_axis, y_axis, label = 'GSP', color='blue', marker = '*', linewidth=4)
# show the legend, without this the label means nothing
plt.legend()

plt.xlabel('Date2')
plt.ylabel('Year2')

plt.title('My second chart')

plt.xlim(0,60)
plt.ylim(0,60)
         
plt.grid()

plt.savefig('my_image2.png')
plt.show()
```
