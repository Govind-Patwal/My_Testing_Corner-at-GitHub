```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python

plt.scatter(x_axis, y_axis, s=[p*3 for p in y_axis], color ="red", label ="Chicago")
plt.legend()
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s= [i*3 for i in y_axis], color ="skyblue", alpha=0.2, edgecolors="black", linewidth=2, label = "boston")
ax.legend()
ax.set_ylabel('Date')
ax.set_xlabel('Fare ($)')
ax.set_title("PyBer Fare by Month")
plt.show()
```
